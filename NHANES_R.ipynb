```r
library(NHANES)
library(nhanesA)
library(tidyverse)
library(arrow)
```

```r
tchol_2013 <- nhanes('TCHOL_H')
tchol_2013
```

```r
tchol_2015 <- nhanes('TCHOL_I')
tchol_2015
```

```r
tchol_2017 <- nhanes('TCHOL_J')
tchol_2017
```

```r
tchol <- left_join(tchol_2013, tchol_2015) %>% left_join(tchol_2017)
```

```r
demo_2013 <- nhanes('DEMO_H')
```

```r
demo_2015 <- nhanes('DEMO_I')
```

```r
demo_2017 <- nhanes('DEMO_J')
```

```r
demo <- left_join(demo_2013, demo_2015) %>% left_join(demo_2017)
```

```r
diet_2013 <- nhanes("DR1TOT_H")
```

```r
diet_2015 <- nhanes("DR1TOT_I")
```

```r
diet_2017 <- nhanes("DR1TOT_J")
```

```r
diet <- left_join(diet_2013, diet_2015) %>% left_join(diet_2017)
```

```r
bp_2013 <- nhanes("BPX_H")
bp_2015 <- nhanes("BPX_I")
bp_2017 <- nhanes("BPX_J")
```

```r
bp <- left_join(bp_2013, bp_2015) %>% left_join(bp_2017)
```

```r
bmi_2013 <- nhanes("BMX_H")
bmi_2015 <- nhanes("BMX_I")
bmi_2017 <- nhanes("BMX_J")
```

```r
bmi <- left_join(bmi_2013, bmi_2015) %>% left_join(bmi_2017)
```

```r
bpq_2013 <- nhanes("BPQ_H")
bpq_2015 <- nhanes("BPQ_I")
bpq_2017 <- nhanes("BPQ_J")
```

```r
physical_2013 <- nhanes("PAQ_H") %>% select(-PAQ724A)
physical_2015 <- nhanes("PAQ_I") %>% select(-PAQ724A)
physical_2017 <- nhanes("PAQ_J")
```

```r
physical <- left_join(physical_2013, physical_2015) %>% left_join(physical_2017)
```

```r
diabetes_2013 <- nhanes("DIQ_H") %>% select(-DIQ175A)
diabetes_2015 <- nhanes("DIQ_I") %>% select(-DIQ175A)
diabetes_2017 <- nhanes("DIQ_J")
```

```r
diabetes <- left_join(diabetes_2013, diabetes_2015) %>% left_join(diabetes_2017)
```

```r
smoking_2013 <- nhanes("SMQ_H") %>% select(-SMQ665D, -SMQ665B, -SMQ665C)
smoking_2015 <- nhanes("SMQ_I") %>% select(-SMQ665D, -SMQ665B, -SMQ665C)
smoking_2017 <- nhanes("SMQ_J")
```

```r
smoking <- left_join(smoking_2013, smoking_2015) %>% left_join(smoking_2017)
```

```r
alcohol_2013 <- nhanes("ALQ_H")
alcohol_2015 <- nhanes("ALQ_I")
alcohol_2017 <- nhanes("ALQ_J")
```

```r
alcohol <- left_join(alcohol_2013, alcohol_2015) %>% left_join(alcohol_2017)
```

```r
combined_df  <- left_join(tchol, demo) %>% left_join(diet) %>% left_join(bp) %>% left_join(bmi) %>%
  left_join(physical) %>% left_join(diabetes) %>% left_join(smoking) %>% left_join(alcohol)
```

```r
sleep_2013 <- nhanes("SLQ_H")
```

```r
sleep_2015 <- nhanes("SLQ_I") %>% mutate(SLD010H = SLD012)
```

```r
sleep_2017 <- nhanes("SLQ_J") %>% mutate(SLD010H = (SLD012*5 + SLD013*2)/7)
```

```r
sleep <- left_join(sleep_2013, sleep_2015) %>% left_join(sleep_2017)
```

```r
glucose_2013 <- nhanes("GLU_H")
```

```r
glucose_2015 <- nhanes("GLU_I")
```

```r
glucose_2017 <- nhanes("GLU_J")
```

```r
glucose <- left_join(glucose_2013, glucose_2015) %>% left_join(glucose_2017)
```

```r
combined_df <- left_join(combined_df, glucose) %>% left_join(sleep)
```

```r
combined_df <- read.csv("Nhanes.csv")
```
